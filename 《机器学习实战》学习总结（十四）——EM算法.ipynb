### 极大似然估计

极大似然估计法是概率论与数理统计中的知识点。比如说我们已知变量Y的概率分布服从高斯分布， $Y\sim N(\mu,\sigma^2)$ ，$ \mu $和 $\sigma $是分布的参数，记 $\theta=(\mu,\sigma)$ 。现在我们想要求这两个参数，怎么做呢？

首先在分布中独立地进行n次采样，将采样样本记作$ y=(y_1,y_2,...,y_n) $，然后我们从这n个样本反推最有可能的参数的值，这个过程就叫做极大似然估计。

记$ P(y_i|\theta)$ 表示在参数为 $\theta $时采样得到样本 $y_i $的概率。因为是独立采样，所以得到在总体样本中取得这n个样本的概率为：
$$
L(\theta)=\prod_{i=1}^{n}P(y_i|\theta)
$$
上面这个函数中， $\theta $是未知的，它表示在不同的$\theta$ 取值下，得到当前这个样本集的可能性。我们想要求的就是一个 $\theta $，使得似然函数 $L(\theta)$ 的取值最大。

一般情况下我们采用对数似然，两边同取对数，得：
$$
H(\theta)=lnL(\theta)=\sum_{i=1}^{n}{lnP(y_i|\theta)}
$$
之后我们令$H(\theta)$ 对$ \theta$ 分别求导等于零，就可以求得参数 $\theta $的取值。

以上就是极大似然估计的内容。



### EM算法

有时候我们的变量不仅仅有可观测到的变量$Y$，还会有隐藏变量$Z$。什么叫隐藏变量呢？

比如说变量 $A\sim N(\mu_1,\sigma_1^2)$ ，变量 $B\sim N(\mu_2,\sigma_2^2) $，要求的参数是 $\theta=(\mu_1,\sigma_1,\mu_2,\sigma_2) $。但这时候$A$、$B$混在了一起，我们从这里面采样得到 $y=(y_1,y_2,...,y_n)$ ，那么$Y$就是可观测变量，但是具体的$ y_i$ 是从$A$中取到的还是B中取到的我们不知道，这就是隐藏变量 $Z=(z_1,z_2,z_3,...z_n)$ ,这里的 $z_i$ 代表$ y_i$ 从A中取得还是从B中取得。

这时候目标是极大化观测数据Y对于参数 \theta 的对数似然函数，即极大化
$$
L(\theta)=logP(Y|\theta)=log\sum_{Z}^{}{P(Y,Z|\theta)}
$$
再由贝叶斯公式，得
$$
L(\theta)=log\sum_{Z}^{}{P(Y,Z|\theta)}=log(\sum_{Z}^{}{P(Y|Z,\theta)P(Z|\theta)})
$$
我们发现在这里我们不能用在极大似然估计法中采用的求导的方法，因为其中含有隐含变量Z。

这时候EM算法就被提了出来，它采用迭代的方法来求解参数 $\theta $，直到参数收敛。

假设我们在第i次迭代后$ \theta$ 的估计值是 $\theta^{(i)} $，我们希望下一次的估计值能够使得 $L(\theta)$ 增加，即
$$
L(\theta)>L(\theta^{(i)})
$$
我们考虑两者作差：
$$
L(\theta)-L(\theta^{(i)})=log(\sum_{Z}^{}{P(Y|Z,\theta)P(Z|\theta)})-logP(Y|\theta^{(i)})
$$
$$
=log\left[ \sum_{z}^{}{}P(Z|Y,\theta^{(i)})\frac{P(Y|Z,\theta)P(Z|\theta)}{P(Z|Y,\theta^{(i)})} \right]-logP(Y|\theta^{(i)})
$$
上式中 $P(Z|Y,\theta^{(i)})$ 是在给定观测数据Y和当前估计的参数 $\theta^{(i)}$ 下隐变量数据Z的条件概率分布， $\sum_{z}^{}{P(Z|Y,\theta^{(i)})}=1 $。

由 $E(f(x))=\sum_{}^{}{P(x)f(x)}$ ，所以上式第一部分
$$
\left[ \sum_{z}^{}{}P(Z|Y,\theta^{(i)})\frac{P(Y|Z,\theta)P(Z|\theta)}{P(Z|Y,\theta^{(i)})} \right]
$$
是 $\frac{P(Y|Z,\theta)P(Z|\theta)}{P(Z|Y,\theta^{(i)})}$  的数学期望。

下面要用到$Jenson$不等式，对于一个凸函数(注：凹凸函数的定义同济版高数教材和国外不太一样，大家不要纠结，理解就行，名字不重要)，有：
$$
f(\sum_{j}^{}{\lambda_jy_j})\geq\sum_{j}^{}{\lambda_jf(y_j)}，其中\lambda_j\geq0,\sum_{j}^{}{\lambda_j=1}
$$
换一种写法，就是
$$
f(E(x))\ge E(f(x))
$$
用图形直观表示如下(来源于网络，侵删)：




所以对上式应用Jensen不等式，得
$$
\begin{aligned} % requires amsmath; align* for no eq. number
&=log\left[ \sum_{z}^{}{}P(Z|Y,\theta^{(i)})\frac{P(Y|Z,\theta)P(Z|\theta)}{P(Z|Y,\theta^{(i)})} \right]-logP(Y|\theta^{(i)})\\
&\geq\sum_{z}^{}{}P(Z|Y,\theta^{(i)})log\frac{P(Y|Z,\theta)P(Z|\theta)}{P(Z|Y,\theta^{(i)})} -logP(Y|\theta^{(i)})\\
&=\sum_{z}^{}{}P(Z|Y,\theta^{(i)})log\frac{P(Y|Z,\theta)P(Z|\theta)}{P(Z|Y,\theta^{(i)})} -\sum_{z}^{}{}P(Z|Y,\theta^{(i)})logP(Y|\theta^{(i)})\\
&=\sum_{z}^{}{}P(Z|Y,\theta^{(i)})log\frac{P(Y|Z,\theta)P(Z|\theta)}{P(Z|Y,\theta^{(i)})P(Y|\theta^{(i)})}
\end{aligned}
$$
令
$$
B(\theta,\theta^{(i)})=L(\theta^{(i)})+\sum_{z}^{}{}P(Z|Y,\theta^{(i)})log\frac{P(Y|Z,\theta)P(Z|\theta)}{P(Z|Y,\theta^{(i)})P(Y|\theta^{(i)})}
$$
所以
$$
L(\theta)\geq B(\theta,\theta^{(i)})
$$
$B(\theta,\theta^{(i)}) $是$ L(\theta) $的一个下界。在这里我们还是有两个参数$Z$和 $\theta$ 。

下面一步就是开创性的一步，也就是$EM$算法中的E步：我们需要先估计出$Z$的值，怎么估计呢，因为
$$
Y=(y_1,y_2,....y_j,...y_n)
$$
$Y$这是我们已知的观测变量
$$
Z=(z_1,z_2,...z_j,...z_n)
$$
$Z$是隐含变量，用我们的例子来说就是代表了$ y_j$ 是从$A$、$B$哪一个分布中中选出的，我们先估计
$$
z_j=argmaxP(z|y_j,\theta^{(i)})
$$
上面公式的意思就是说，我们根据已知的 $\theta^{(i)} $，找出对应可观测变量 $y_j$ 最有可能的 $z_j $,把其作为这一次迭代的估计值。

这样通过上面的$E$步，我们就有了这一轮Z的估计值。

这就是E步，在求得 Z 的估计值之后，再通过优化 $B(\theta,\theta^{(i)}) $，从而提高下界来优化$ L(\theta)$ ：

$$
\theta^{(i+1)}=argmax\left[ L(\theta^{(i)})+\sum_{z}^{}{}P(Z|Y,\theta^{(i)})log\frac{P(Y|Z,\theta)P(Z|\theta)}{P(Z|Y,\theta^{(i)})P(Y|\theta^{(i)})} \right]
$$

忽略其中对$ \theta$ 而言是常数项的部分，得
$$
\begin{aligned} % requires amsmath; align* for no eq. number
&=argmax\left[ \sum_{z}^{}{}P(Z|Y,\theta^{(i)})log(P(Y|Z,\theta)P(Z|\theta)) \right]\\
&=argmax\left[ \sum_{z}^{}{}P(Z|Y,\theta^{(i)})logP(Y,Z|\theta) \right]\\
&=argmax(Q(\theta,\theta^{(i)}))
\end{aligned}
$$
这是$M$步。

在求得 $\theta^{(i+1)}$ 后，继续进行$E$步：令 $\theta=\theta^{(i+1)} $，求$Z$的估计值。反复迭代，直到结果$ \theta$ 收敛。



### EM算法的收敛性

现在我们希望知道经过反复迭代， $\theta$ 的值到底会不会收敛。

首先证明 $L(\theta^{(i+1)})\geq L(\theta^{(i)})$ 。

因为
$$
L(\theta^{(i+1)})-L(\theta^{(i)})\geq B(\theta^{(i+1)},\theta^{(i)})-B(\theta,\theta^{(i)})
$$
且
$$
\theta^{(i+1)}=argmaxB(\theta,\theta^{(i)})
$$
所以
$$
B(\theta^{(i+1)},\theta^{(i)})-B(\theta,\theta^{(i)})\geq0
$$
所以
$$
L(\theta^{(i+1)})-L(\theta^{(i)})\geq0
$$
所以说 $L(\theta)$ 是一个单调递增函数。

又因为
$$
L(\theta)=logP(Y|\theta)
$$
且$ P(Y|\theta) $有界，所以$L(\theta) $会收敛到某一个点 $L^*(\theta) $。也即 $\theta$ 会收敛到某一点$\theta^*$ 。

以上我们证明了$EM$算法的收敛性。但我们后续发现EM算法只能收敛到局部最优点，很难找到全局最优点。为了解决这个问题，我们会通常选取多个初值 $\theta^{(0)}$ 多次训练，然后对得到的估计值加以比较，找到最好的。也就是说EM算法对初值的选取比较敏感。



以上就是$EM$算法的全部内容。
